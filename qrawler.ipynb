```python
#perform operations:open chrome
def open_chrome():
    check_internet_connection()
    print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Opening chrome...{Fore.RESET}\n", end="")
    retry_delay=2
    for retry_count in range(1, max_retries+1):
        try:
            # Set up the ChromeDriver service
            service = Service(ChromeDriverManager().install())
            # Use the service and ChromeOptions to instantiate the Chrome driver
            options = Options()
            driver = webdriver.Chrome(service=service, options=options)
            print(f"{Fore.GREEN}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.GREEN} Chrome opened successfully!{Fore.RESET}\n", end="")
            print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Redirecting to verbit.com ...{Fore.RESET}\n",end="")
            has_chrome_opened = True
            login(driver)
            break
        except Exception as e:
            # An exception occurred while opening Chrome
            print(Fore.RED + f"[+] {datetime.now().strftime('%H:%M:%S')}-> Failed to open Chrome\n", end="")
            if retry_count < max_retries:
                print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Waiting {retry_delay} seconds before retrying...{Fore.RESET}\n", end="")
                time.sleep(retry_delay)
                retry_delay *= 2  # Increase the retry delay exponentially
            else:
                print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Maximum number of retries reached.{Fore.RESET}\n", end="")
                print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Unable to open Chrome.{Fore.RESET}\n", end="")
                print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Shutting down...{Fore.RESET}\n", end="")
                break
```

```python
#claim job
def claim_job(driver):
    print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Fetching jobs...{Fore.RESET}\n",end="")
    start_keyboard_listener(driver)
    while True:
        time.sleep(refresh_interval)
        driver.refresh()
        
    
        #wait for links
        wait=WebDriverWait(driver,refresh_interval)
        link_elements=wait.until(EC.presence_of_all_elements_located((By.XPATH,"//a[@class='link-class']")))

        if link_elements:
            #randomize the links
            random.shuffle(link_elements)
        
            #open each new link in a new tab
            for link in link_elements:
                if os_name == "Windows":
                    link.send_keys(Keys.CONTROL + Keys.RETURN) # For windows
                elif os_name == "Linux":
                    link.send_keys(Keys.CONTROL + Keys.RETURN) # For Linux
                elif os_name == "Darwin":
                    link.send_keys(Keys.COMMAND + Keys.RETURN) # For MACOS
                else:
                    pass
                
                #switch to the new tab
                driver.switch_to.window(driver.window_handles[-1])
                
                
                #generate delay
                delay=random.uniform(min_delay, max_delay)
                
                #switch back to the original tab
                driver.switch_to.window(driver.window_handles[0])
                
                #generate another delay 
                delay=random.uniform(min_delay, max_delay)
                
                time.sleep(delay)
                
            #wait  for specified refresh interval
            driver.implicitly_wait(refresh_interval)
```

```python
#keyboard listener
def keyboard_listener(driver):
    global should_quit,driver
    while True:
        key = input("Press Q/q to quit the script.\n")
        if key.lower() == 'q':
            should_quit = True
            break
```

```python
#start keyboard listener
def start_keyboard_listener(driver):
    keyboard_thread = threading.Thread(target=keyboard_listener(driver))
    keyboard_thread.daemon = True
    keyboard_thread.start()
```

```python
def login(driver):
    #open verbit login page
    driver.get(url)
    driver.maximize_window()
    print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Attempting to login...{Fore.RESET}\n",end="")
    #locate and fill the login form
    try:
        #explicit wait
        wait=WebDriverWait(driver,10)
        print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Inputting email...{Fore.RESET}\n",end="")
        email_field = driver.find_element(By.XPATH, "//input[@name='email']")
        email_field.send_keys(email)
        print(f"{Fore.GREEN}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.GREEN} Done.{Fore.RESET}\n",end="")
        # Locate the button by its text using XPath
        next_button = driver.find_element(By.XPATH, "//button[contains(text(),'Next')]")

        # Click on the button
        next_button.click()

       
        password_field=wait.until(EC.visibility_of_element_located((By.XPATH,"//input[@type='password']")))
        print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Inputting password...{Fore.RESET}\n",end="")
        password_field.send_keys(password)
        print(f"{Fore.GREEN}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.GREEN} Done.{Fore.RESET}\n",end="")
        print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.BLUE} Authenticating...{Fore.RESET}\n",end="")
        #Locate the button by its text using XPath
        login_button = driver.find_element(By.XPATH, "//button[contains(text(),'Login')]")
        #Click on the button
        login_button.click()
        
        #error = wait.until(EC.visibility_of_element_located((By.TAG_NAME, "p")))
        try:
            login = wait.until(EC.visibility_of_element_located((By.XPATH, "//h1[@class='title' and contains(text(), 'My files')]")))
            if login:
                print(f"{Fore.GREEN}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.GREEN} Logged in successfully.{Fore.RESET}\n",end="")
                claim_job(driver)
        except NoSuchElementException as e:
           pass

    except Exception as e:
        page_source=driver.page_source
        verification_text="Enter verification code"
        error_text="Incorrect username or password"
        #verification_text = wait.until(EC.visibility_of_element_located((By.XPATH, "//h1[@class='fWwCIu' and contains(text(), 'Enter verification code')]")))
        #error = wait.until(EC.visibility_of_element_located((By.TAG_NAME, "p")))
        if verification_text in page_source:
            print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.YELLOW} Kindly login in some other tab before using this script...{Fore.RESET}\n",end="")
        elif error_text in page_source:
            print(f"{Fore.MAGENTA}[+]{datetime.now().strftime('%H:%M:%S')}->{Fore.RESET}{Fore.RED} Authentication failed.Either email address provided or password is incorrect.{Fore.RESET}\n",end="")
        start_keyboard_listener(driver)
```
